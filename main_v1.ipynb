###1.1 Write a function that reurns Count number of lines in txt file

```python
def NumOfLines(file_name):
    path = r"../input/text-file-txt/" + file_name
    
    with open(path, "rb") as file:
        # readlines() returns a list of lines and each element ends with \r\n 
        lines = file.readlines()
        count = len(lines)
    
    return count
```

```python
FILE_NAME = r"text_file.txt" 
print("Number of Lines are ", NumOfLines(FILE_NAME))
```

###1.2 Write a function that reurns Count number of words in txt file

```python

def NumOfWords(file_name):
    path = r"../input/text-file-txt/" + file_name
    words = 0
    
    with open(path, "rb") as file:
        lines = file.readlines()
        for line in lines:
            words += len(line.split())
        
    return words
```

```python
FILE_NAME = r"text_file.txt"
print("Number of words are ", NumOfWords(FILE_NAME))
```

##Task 2
Write a function that reurns a dataframe with two columns having stop words in one col and their count in the other.

```python
import pandas
from nltk.corpus import stopwords


def CreateDF(file_name):
    # List containing all English stopwords from NLTK
    NLTK_STOP_WORDS = stopwords.words('english')
    
    # Converting each word from String type to Bytes
    NLTK_STOP_WORDS = [bytes(word, 'utf-8') for word in NLTK_STOP_WORDS]
    
    data = {
        "stop_words" : [], 
        "count" : []
    }
    
    path = r"../input/text-file-txt/" + file_name
    
    # Dictionary initialized with stopwords as key and 0 as values
    count = {key: 0 for key in NLTK_STOP_WORDS}
    
    with open(path, 'rb') as file:
        lines = file.readlines()
        for line in lines:
            for word in line.split():
                if word in NLTK_STOP_WORDS:
                    count[word] += 1
                    
    for key, value in count.items():
        if value > 0:
            # Converting key from Bytes type to String type
            key = str(key, 'utf-8')
            data["stop_words"].append(key)
            data["count"].append(value)

    data_frame = pandas.DataFrame(data)
    return data_frame
```

```python
FILE_NAME = r"text_file.txt"
data_frame = CreateDF(FILE_NAME)
print(data_frame.to_string())
```
